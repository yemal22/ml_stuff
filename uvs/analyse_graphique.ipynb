```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
plt.style.use('ggplot')
```

```python
tips_df = sns.load_dataset('tips')
tips_df.head()
```

### Histogramme

```python
# Histogramme du montant total de la consommation

sns.displot(tips_df['total_bill'], kde=False).set_title('Consommation totale')
```

```python
# Histogramme du montant total de la consommation

sns.displot(tips_df['total_bill'], kde=False, bins=50)
```

```python
# Densité

sns.displot(tips_df['total_bill'], kde=True)
```
