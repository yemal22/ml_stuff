```python
from knn import KNN
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.model_selection import train_test_split
from matplotlib.colors import ListedColormap
import numpy as np
```

```python
iris = datasets.load_iris()
X, y = iris.data, iris.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1234)
```

```python
cmap = ListedColormap(["#FF0000", "#00FF00", "#0000FF"])
```

```python
plt.figure()
plt.scatter(X[:,2], X[:,3], c=y, edgecolors='k', cmap=cmap)
```

```python
knn = KNN(k=5)
```

```python
knn.fit(X_train, y_train)
```

```python
print(knn.predict(X_test))
```

```python
predictions = knn.predict(X_test)
```

```python
(predictions == y_test).mean()
```

```python
train_accuracy = []
test_accuracy = []
for i in range(1, 11):
    clf = KNN(k=i)
    clf.fit(X_train, y_train)
    train_accuracy.append((clf.predict(X_train) == y_train).mean())
    test_accuracy.append((clf.predict(X_test) == y_test).mean())
```

```python
plt.figure()
plt.plot(range(1, 11), train_accuracy, marker='o', color='blue', label='Train Accuracy')
plt.plot(range(1, 11), test_accuracy, marker='o', color='green', label='Test Accuracy')
plt.xlabel("K")
plt.ylabel("Accuracy")
plt.title("Predictions accuracies on train data vs test data")
plt.legend()
```
