```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
prix_ms = [93, 50, 57, 69, 94, 100]
prix_apple = [93, 112, 104, 144, 170, 172]
annee = [2015, 2016, 2017, 2018, 2019, 2020]
```

```python
plt.subplot(1, 2, 1)
plt.bar(annee, prix_ms)
plt.title("Prix Windows")
plt.subplot(1, 2, 2)
plt.bar(annee, prix_apple)
plt.title("Prix Apple")
```

```python
plt.figure()
plt.bar(annee, prix_ms, label='Prix Windows')
plt.bar(annee, prix_apple, label='Prix Apple', bottom=prix_ms)
plt.legend()
plt.xlabel('annee')
plt.ylabel('Prix')
```

```python
plt.figure()
annee_m = np.array(annee)
a = 0.25
plt.bar(annee_m, prix_ms, label='Prix Windows', width=0.25)
plt.bar(annee_m + a, prix_apple, label='Prix Apple', width=0.25)
plt.legend()
plt.xlabel('annee')
plt.ylabel('Prix')
```
